```python
import os
import math
os.chdir("/root/dev/tvm/pim_test/upmem/baseline/GEMV_SAMSUNG/")

def run(m, n, dpus, ntile, test):
    bl = max(5, int(math.log2(m / ntile) + 2))
    tasklet_mult = n * ntile / 2 / dpus
    tasklets = 1
    for i in range(16, 1, -1):
        if tasklet_mult % i == 0:
            tasklets = i
            break
    if test:
        print(f"bl={bl}")
        print(f"tasklets={tasklets}")

        os.system(f"make clean")
        os.system(f"NR_DPUS={dpus} NR_TASKLETS={tasklets} TYPE=INT32 BL={bl} make 2> /dev/null")
        os.system(f"./bin/gemv_host -m {m} -n {n} -w 1 -e 3 -N {ntile}")
    else:
        os.system(f"make clean")
        os.system(f"NR_DPUS={dpus} NR_TASKLETS={tasklets} TYPE=INT32 BL={bl} make 2> /dev/null")
        os.system(f"./bin/gemv_host -m {m} -n {n} -w 10 -e 1000 -N {ntile}")
```

```python
run(8192, 8192, 2048, 1, True)
run(8192, 8192, 2048, 2, True)
run(8192, 8192, 2048, 4, True)
run(8192, 8192, 2048, 8, True)
run(8192, 8192, 2048, 16, True)
```
